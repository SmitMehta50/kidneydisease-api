```python
import tflite_runtime.interpreter as tflite
```

```python
save_path = '/content/drive/MyDrive/LY Project/Kidney disease'
tflite_model_name = 'kidney_prediction_model.tflite'
tflite_model_path = save_path + '/' + tflite_model_name

interpreter = tflite.Interpreter(model_path = tflite_model_path)
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
print("Input Shape:", input_details[0]['shape'])
print("Input Type:", input_details[0]['dtype'])
print("Output Shape:", output_details[0]['shape'])
print("Output Type:", output_details[0]['dtype'])
```

```python
import joblib
import os

os.chdir(save_path)
scaler = joblib.load('scaler.joblib')
```

```python
col = ['age',
 'blood_pressure',
 'specific_gravity',
 'albumin',
 'sugar',
 'red_blood_cells',
 'pus_cell',
 'pus_cell_clumps',
 'bacteria',
 'blood_glucose_random',
 'blood_urea',
 'serum_creatinine',
 'sodium',
 'potassium',
 'haemoglobin',
 'packed_cell_volume',
 'white_blood_cell_count',
 'red_blood_cell_count',
 'hypertension',
 'diabetes_mellitus',
 'coronary_artery_disease',
 'appetite',
 'pedal_edema',
 'anemia']
```

```python
d = {}
d['yes'] = 1
d['no'] = 0
d['present'] = 1
d['notpresent'] = 0
d['abnormal'] = 1
d['normal'] = 0
d['good'] = 1
d['poor'] = 0

d
```

```python
def model_prediction(data):
    interpreter.allocate_tensors()
    interpreter.set_tensor(input_details[0]['index'], data)
    interpreter.invoke()
    pred = interpreter.get_tensor(output_details[0]['index'])
    return pred
```

```python
def predict(data):
    for i in range(len(data)):
        try:
            data[i] = float(data[i])
        except:
            data[i] = d[data[i]]
    data = scaler.transform([data])
    data = data.astype('float32')
    prob = model_prediction(data)[0][0]

    if round(prob) == 0:
        return round(prob),1-prob
    else:
        return round(prob),prob
```

```python
age = 54                               #int
blood_pressure = 70                    #int
specific_gravity = 1.005               #float
albumin = 4                            #int
sugar = 0                              #int           
red_blood_cells = 'abnormal'           #'abnormal' or 'normal'
pus_cell = 'normal'                    #'abnormal' or 'normal'
pus_cell_clumps = 'notpresent'         #'present' or 'notpresent'
bacteria = 'present'                   #'present' or 'notpresent'
blood_glucose_random = 117             #int
blood_urea = 56                        #int
serum_creatinine = 3.8                 #float
sodium = 111                           #int
potassium = 2.5                        #float
haemoglobin = 11.2                     #float
packed_cell_volume = 32                #int
white_blood_cell_count = 6700          #int
red_blood_cell_count = 3.9             #float
hypertension = 'yes'                   #'yes' or 'no'
diabetes_mellitus = 'yes'              #'yes' or 'no'
coronary_artery_disease = 'yes'        #'yes' or 'no'
appetite = 'poor'                      #'poor' or 'good'
pedal_edema = 'no'                     #'yes' or 'no'
anemia = 'no'                          #'yes' or 'no'

data = [eval(i) for i in col]
pred,prob = predict(data)
```
